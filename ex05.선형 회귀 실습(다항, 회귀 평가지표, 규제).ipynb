````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# 데이터 셋 불러오는 코드
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
# 정답데이터
target = raw_df.values[1::2, 2]
````

````python
from scipy.stats import shapiro, kstest, norm
# scipy : 검정 도구
# shapiro : 샤피로 윌크, 데이터 크기가 50 이하일 때 정규분포를 따르는지 검정하는 도구
# kstest : 콜모고로프-스미르노프, 큰 표본에서 정규분포를 따르는지 검정하는 도구
# norm : 정규분포를 다루는 모듈

# 정규성 테스트 결과 저장
results = {}
for column in boston.columns:
     # 각 열에 접근하여 정규분포를 비교하여 stat(테스트 통계량), p(p-value)
    stat, p = kstest(boston[column], 'norm', args = (boston[column].mean(), boston[column].std()))  
    results[column] = (stat, p)

# 결과 출력
for column, (stat, p) in results.items():
    print(f'{column} - Statistics={stat}, p-value={p}')
    if p > 0.05:   # 기각하냐 기각하지 않냐 를 물어보는 것
        print(f'{column}는 정규분포를 따릅니다.')
    else:
        print(f'{column}는 정규분포를 따르지 않습니다.')
        
# => 정규분포를 따르지 않으니 비정규분포 => MinMaxScaler 불러오기
````

````python
# 거리/수치 기반 알고리즘의 경우 특정 컬럼이 영향을 많이 주지 않게끔 스케일링 작업 필요! 
from sklearn.preprocessing import MinMaxScaler   # 0 ~ 1 사이의 값으로 범위 조정
````

````python
# 스케일러 초기화
mm_scaler = MinMaxScaler()

X_scaled = mm_scaler.fit_transform(boston)   # 기준 잡고 바꿔버리기!
# fix : 기준 잡는다
# 주의사항 : 나중에 train, test 나눴을 때 train에 fix이나 fix_transform을 적용시켰다면, test에는 transform만 적용 시켜야 함!
X_scaled
````

````python
X_scaled = pd.DataFrame(X_scaled, columns = boston.columns)
````

````python
X_scaled
# => 0 ~ 1 사이의 값으로 전부 반환시킴
````

````python
# 순서대로 네 개의 데이터 가져와서
X_train, X_test, y_train, y_test = train_test_split(X_scaled, target,
                                                    test_size = 0.3, 
                                                    # 랜덤한 값 고정시키기(올해가 2024년이라 그냥 2024로 한 것)
                                                    random_state = 2024)
````

### Linear Model - 회귀 평가 지표
- MSE(평균 제곱 오차, Mean Squared Error) : 실제값과 예측값 차이를 제곱해서 평균낸 것
  - 많이 쓰는 이유 : 큰 오차 값에 대해 더 큰 숫자값을 부여함 => 오차의 정보를 극대화시킴
  - 단점 : 1 미만의 정보는 오차의 정보 감소됨 -> 그때는 MAE 쓰기
- MAE(평균 절대 오차, Mean Absoute Error) : 실제값과 예측값 차이의 절대값을 평균낸 것
  - MSE의 한계로는 1 미만의 오차값들은 오차의 정보가 손실되는 상황이 발생!
  - 0 ~ 1 사이의 오차값을 가지는 경우 유용하게 사용 가능(특이한 케이스에서 사용)
- RMSE(평균 제곱근 오차, Root MSE) : MSE에 Root 씌운 것
  - 오차의 단위를 원래 단위로 변환
  - MSE로 했을 때 오차값이 너무 크면 유용하게 사용 가능
  - MAE보다 작은 단위로 하고 싶을 때?
- R-squared(r2score, 결정계수) : 모델(예측값)이 데이터의 퍼짐 정도(분산, 변동성)를 얼마나 이해하고 있는지를 나타내는 지표값
  - 0 ~ 1 사이의 값으로 표현, 1에 가까울수록 모델이 데이터를 잘 이해하고 있다를 의미(모델이 데이터를 잘 설명하고 있다)
---


- 예를 들어 MSE에서 오차값이 0.5가 나왔을 경우 제곱하면 0.25가 되는데 오차값이 점점 떨어짐 => 이럴 때에는 MAE나 RMSE를 쓰자 ?
---


````python
# 회귀 평가 지표 도구들 불러오기
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
````

````python
# 모델 초기화 - 예측값(y_test) 꼭 필요
lr_rg = LinearRegression()

# 학습
lr_rg.fit(X_train, y_train)

# 예측
y_pred = lr_rg.predict(X_test)
````

````python
# X_test에 대한 예측값
y_pred
````

````python
# MSE 확인(실제값, 예측값) - 헷갈리면 shift + Tap 눌러서 뭘 넣어야하는지 확인하기
mean_squared_error(y_test, y_pred)
````

````python
# MAE 확인(오차의 절대값을 평균낸 값) - 0 ~ 1
mean_absolute_error(y_test, y_pred)
````

````python
# RMSE 확인(**0.5 => Root) - MAE의 오차값이 너무 클 때 사용
mean_squared_error(y_test, y_pred) ** 0.5
# => 오차값이 크면 좋기에 MAE 많이 사용
````

````python
# r2score 확인 (0.8 이상이면 좋다 - 1에 너무 가까우면 과대적합 의심)
r2_score(y_test, y_pred)
````

````python
# 경사하강법이 적용된 모델로도 평가 지표를 확인해서 성능 비교하기
from sklearn.linear_model import SGDRegressor

# 모델 초기화
sgd_rg = SGDRegressor(
    eta0 = 0.01,   # 학습률
    max_iter = 10000,   # 최대 학습 빈도수
    verbose = 1   # 학습 상황 출력(0으로 설정하면 출력되지 않음)
)

# 학습
sgd_rg.fit(X_train, y_train)

# 예측
y_pred = sgd_rg.predict(X_test)

# => loss값(MSE값) 중요시 보기 -> Epoch 292 수정했다!
# NNZs : 0이 아닌 w값(x 옆에 붙어있는 가중치) 개수
# T : 연산량(횟수)
````

````python
print(f"평균 제곱 오차 : {mean_squared_error(y_test, y_pred)}")
print(f"R^2 : {sgd_rg.score(X_test, y_test)}")
# R^2는 {r2_score(y_test, y_pred)}로 써줘도 값이 똑같이 나옴

# 성능이 꽤 괜찮지는 않은 듯 함 => 이유는? 모델이 너무 단순해서
````

````python
# 기준 설정(2차항으로 기준 잡아줌)
poly.fit(X_scaled)
````

````python
X_train_poly = poly.transform(X_train)   # 변환
````

#### 규제(Regularization)
- w(가중치)가 크다면, X(Feature)가 조금만 달라져도 y값은 크게 변함
- w에 억제를 가할 수 있음!
---
1. L2 규제(Ridge, SGD의 기본값) : 모든 w에 적절한 힘으로 규제를 적용하는 방법
   - 모든 feature(특성)을 y를 예측하는데 사용되게끔 규제를 가함 → 특정 계수(w값)들은 0이 되지 않음
2. L1 규제(Lasso) : 모든 w에 완전히 똑같은 힘으로 규제를 적용하는 방법
   - 특성 선택(Feature Selection)이 일어남 → 특정 계수들은 0이 되어버림

````python
# L1 규제 가한 것!
# SGDRegressor 모델 학습, 평가
sgd_rg =  SGDRegressor(
    eta0 = 0.06, # 학습률 설정
    max_iter = 10000,
    verbose = 1,
    random_state=2024, # 시드(난수) 고정
    penalty = 'l1', #Lasso 규제, 특정 w값이 0이 되게끔 규제를 가함 (Feature Selection)
    alpha = 0.002  # 규제 강도, 높으면 높을 수록 규제가 강해짐!
    
)
sgd_rg.fit(X_train_poly, y_train) # 확장된 문제 데이터로 학습
y_pred = sgd_rg.predict(X_test_poly)


# 0이 아닌 w값이 104개였는데 라쏘를 적용시키고 확인해보니 특정 컬럼 (= w, 문제 데이터)들 일부는 꺼진 상태 (0이 됨) 
# => 특성 선택이 일어난 것
````

````python
print(f"평균 제곱 오차 : {mean_squared_error(y_test, y_pred)}")
print(f"R^2 : {sgd_rg.score(X_test_poly, y_test)}")
````

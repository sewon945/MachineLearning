### Text Mining vs NLP
- Text Mining : 비정형 데이터(텍스트)에서 의미 있는 정보를 추출하는 방법
- NLP(자연어 처리) : 사람이 사용하는 언어를 이해하고 예측하는 분야

#### Text Mining 분야
- 문서 분류 : text가 특정 카테고리에 속하는지 예측하는 분야
- 문서 요약 : text의 주요 내용을 추리는 분야
- 문서 군집분석 : 문서(text)끼리의 유사도를 비교 분석하는 분야(ex. 추천알고리즘)
- 감성분석 : text에 들어있는 감정이나 의도를 분석하는 분야

#### Text Minig 프로세스
1. 텍스트 전처리
   - 클렌징(대/소문자, 특수문자 제거, 오탈자 제거 등)
   - 불용어 제거(stop words)
   - 어근 추출(Stemming Lemmatization)
     - 영어 : NLTK
     - 한국어 : Konlpy(형태소 분석기)
   - 토큰화 : text를 일정단위(ex.띄어쓰기, 형태소, 2-gram 등)로 분해하는 작업
2. 수치화/Feature Vector 화 : 토큰화된 토큰들을 가지고 일정 규칙에 따라 숫자를 변경하는 작업
   - 빈도 기반의 수치화(BOW, TF-IDF 등)
   - Word Embedding
3. ML/DL 학습 -> 예측 -> 평가, 분석

#### 단어 빈도 분석(워드 카운팅)

````python
text_train = train["문장"]
````

````python
# 토큰화 작업
tmp = [doc.split(" ") for doc in text_train]
#text_train[0].split(" ") => doc.split(" ")
# => 띄어쓰기를 기준으로 토큰화 수행 (리스트로 바꿈?)
````

### 데이터 클렌징
- 한 글자 단어 삭제 -> 텍스트 마이닝 단계에서는 의미 부여가 어려움
- ㅋㅋㅋ, ㅎㅎㅎ 같은 반복적인 글자 삭제

````python
import re   # 정규표현식, Regular Expression
````

````python
p = re.compile("[ㅋㅎㄷ~ㅡ>!]+")   
# 정규표현식 패턴 : 특정 문자열의 규칙을 가지는 문자열의 집합을 표현하는 데 사용되는 언어) - chatGPT한테 물어보면 알아서 짜줌
````

````python
인종_국적_token
````

```python
# 어떠한 요소가 정규표현식 패턴에 감지되어 있는지 인덱싱 번호 꺼내오기
for i, token in enumerate(인종_국적_token) :
# 문자열 내부에 ㅋㅋ이라는 문자가 들어가 있냐?
    if 'ㅋㅋ' in token :
        print(i)
        
 => 인종_국적_token에서 정규표현식이 감지된 것 - 48번
```

````python
p.match(인종_국적_token[48])
````

````python
if p.match(인종_국적_token[48]) :
    print("True였음!")
````

````python
p.match(인종_국적_token[0])
# => 정규표현식이 감지되지 않으면 아무것도 뜨지 않음
````

````python
인종_국적_token
````

````python
clean_text = []   # 클렌징한 단어들이 들어갈 리스트

for token in 인종_국적_token :
    if len(token) < 2 :  # 문자열(글자) 개수가 2 미만인가? 예수 -> 십새끼 -> 개새끼 -> ...
        continue   # 글자 개수가 2 미만인 경우에는 continue를 만나서 아래쪽 작업을 수행하지 않고 다음 요소를 들여다보라고 다시 위로 올라감
    elif p.match(token) :   # 정규표현식에 매칭되는가? ㅋㅋ은 정규표현식에 매칭돼서 다시 올라감
        continue
    clean_text.append(token)   # 나머지 정상 토큰을 리스트에 쌓기
````

#### 감성분석
- 혐오표현 : 0, 악플/욕설 : 1, clean : 2 로 나눔

````python
train["혐오표현"] = train.loc[:,"여성/가족":"기타 혐오"].sum(axis = 1)
# 여성/가족	남성, 성소수자, 인종/국적, 연령, 지역, 종교, 기타 혐오 다 모아서 "혐오표현"으로 다시 만듦
````

````python
train2 = train[["문장", "혐오표현", "악플/욕설", "clean"]]   # 4가지 컬럼 인덱싱
````

````python
def label(row) :  # axis를 1로 설정할 경우 각 행들이 Series 형으로 넘어옴 - row(행)라는 매개변수로 받아줌
    if row["혐오표현"] >= 1 :   # 혐오표현이 1 이상이냐?
        return 0
    elif row["악플/욕설"] >= 1 :
        return 1
    else :   # clean한 댓글
        return 2
````

````python
from tqdm import tqdm as tq
# 타카둠 : 반복문의 진행 상황을 출력하는 도구
````

````python
train_morphs = []
for txt in tq(text_train) :
    morphs = okt.morphs(txt)
    train_morphs.append(morphs)
````

````python
test_morphs = []
for txt in tq(test2["문장"]) :
    morphs = okt.morphs(txt)
    test_morphs.append(morphs)
````

````python
train_morphs
````

````python
# 정규표현식 패턴 생성
p = re.compile("[ㅋㅎㄷ~!.?,0-9a-zA-Z]+")   # 대괄호 옆에 +를 붙여줘야 2글자 이상을 의미함
````

````python
clean_text_train = []
for sentence in tq(train_morphs) :   # 한 문장씩 꺼내오는 반복문 , # 1차원인 train_morphs를 sentence로 접근
    tmp = []
    for token in sentence :   # 형태소 단위로 쪼개진 token을 꺼내오는 반복문 , # token에 한 번 더 접근
        if len(token) < 2 :  # 한 글자 이하냐?
            continue
        elif p.match(token) :
            continue
        tmp.append(token)   # 클렌징한 토큰들만 담아주기
    clean_text_train.append(tmp)   # 2차원으로 다시 담아주기
````

````python
clean_text_test = []
for sentence in tq(test_morphs) :   # 한 문장씩 꺼내오는 반복문 
    tmp = []
    for token in sentence :   # 형태소 단위로 쪼개진 token을 꺼내오는 반복문
        if len(token) < 2 : 
            continue
        elif p.match(token) :
            continue
        tmp.append(token)  
    clean_text_test.append(tmp) 
````

````python
# 1차원 형태가 담겨있는 2차원 배열
clean_text_test
````

### 수치화(벡터화)
- BOW(Bag of words) : 문장에서 특정 단어의 등장 빈도수를 수치화하는 방법

````python
from sklearn.feature_extraction.text import CountVectorizer   # BOW 클래스 불러오기
````

````python
# 벡터와 수행하기 전에 맛보기로 샘플데이터 생성! - 문장들을 감싸고 있는 리스트들이 하나 씩 더 있기에 2차원배열
sample_text = ["나는 오늘 점심에 삼계탕을 먹었다 왜",
               "너 저녁에 운동갈꺼야?",
               "나는 오늘 저녁에 운동하러 갈꺼야",
               "좋겠다. 나는 저녁에 야간 자율 학습 하러 갈건데..."]
````

````python
sample_cv = CountVectorizer()   # BOW 객체 생성(초기화)
````

````python
# 단어 사전 구축

# 기본값으로는 띄어쓰기 단위로 토큰화 수행
# 하나의 단어는 등록하지 않음(ex. 왜)
sample_cv.fit(sample_text)
````

````python
# 단어 사전 확인
sample_cv.vocabulary_

# key : 각각의 단어 사전
# value : 빈도수 아님 -> 컬럼번호(인덱스번호)
````

````python
# 단어 사전을 기반으로 빈도를 세서 수치화
result = sample_cv.transform(sample_text)
````

````python
result
````

````python
pd.Series(sample_cv.vocabulary_).sort_values().index
# .sort_values() : 인덱스 번호순대로 나열
````

````python
sample_text
````

````python
# 판다스 DF만들어 준 
pd.DataFrame(result.toarray(), columns = pd.Series(sample_cv.vocabulary_).sort_values().index)
````

##### 실제데이터가 Vectorize 적용!

````python
unsmile_cv = CountVectorizer(
    stop_words = ["으로", "이다", "하고", "부터"],   # 불용어 등록
    ngram_range = (1, 2),   # bi-gram 설정
    max_df = 0.9,   # 최대 등장 빈도, 0 ~ 1 or int, 기본값 1, 지정된 인계값(0.9/90%)보다 높은 빈도의 경우 단어 사전 무시
    min_df = 10  # 최소 등장 빈도, 0 ~ 1 or int, 기본값 1, 지정된 인계값(0.9/90%)보다 낮은 빈도의 경우 단어 사전 무시
)

# stop_words : 리스트 안에 들어가는 건 단어 사전으로 채택하지 않겠다
# ngram_range = (1, 1) : 하나의 단어만 쓰겠다
# max_df : 최대 등장 빈도수 제어
````

````python
# CountVectorizer의 입력값은 ID(요소는 문자열)
# 2D 데이터를 띄어쓰기 단위로 다시 문자열을 요소로 가지는 1D로 바꿔줌

# " ".join(clean_text_train[3])
clean_text_train2 = [" ".join(txt) for txt in clean_text_train]
clean_text_test2 = [" ".join(txt) for txt in clean_text_test]
````

````python
clean_text_train2[:5], clean_text_test2[:5]
# => 리스트가 벗겨지고 하나의 문자열로 변경
````

````python
unsmile_cv.fit(clean_text_train2 + clean_text_test2)   # 병합본으로 단어 사전 구축
````

````python
unsmile_cv.vocabulary_   # 단어 사전 확인
# => 수행하면 딕셔너리 형태로 단어와 컬럼명으로 빈도수로 나타남
````

````python
len(unsmile_cv.vocabulary_)   # 딕셔너리 길이 확인
````

````python
# 단어 사전 기반으로 수치화 수행
trans_X_train = unsmile_cv.transform(clean_text_train2)
trans_X_test = unsmile_cv.transform(clean_text_test2)
````

````python
trans_X_train
````

#### 모델링

````python
from sklearn.linear_model import LogisticRegression   # 선형 분류 모델
from sklearn.model_selection import cross_val_score   # 검증용 도구
````

````python
lr_clf = LogisticRegression(max_iter = 10000)   # 모델 초기화(생성)
````

````python
cross_val_score(lr_clf, trans_X_train, y_train, cv = 5).mean()
````

````python
# 하이퍼 파라미터 튜닝 + 하나의 기능으로도 묶어보기!
from sklearn.model_selection import GridSearchCV   # 하이퍼파라미터 튜닝 도구 불러오기
from sklearn.pipeline import Pipeline   # 하나의 기능으로 묶어주는 도구
````

````python
unsmile_pipeline = Pipeline(
    [("unsmile_cv", CountVectorizer(stop_words = ["으로", "이다", "하고", "부터"])),
     ("unsmile_logi", LogisticRegression(max_iter = 10000))]
)
````

````python
unsmile_pipeline
````

````python
# 하이퍼파라미터 경우의 수 만들기
grid_params = {
    "unsmile_cv__max_df" : [0.9, 0.85, 0.8],
    "unsmile_cv__min_df" : [10, 20, 30, 40],   # 단어 사전에 등록될 최소 커트라인
    "unsmile_cv__ngram_range" : [(1, 1), (1, 2), (1, 3)],
    "unsmile_logi__C" : [1000, 100, 10, 1, 0.1, 0.01, 0.001]   # 오른쪽으로 주제?가 갈 수록 높아진다
}
````

````python
grid = GridSearchCV(
    unsmile_pipeline, 
    grid_params,
    cv = 3,
    n_jobs = -1   # 모든 자원을 가져다 써서 찾아내라 라는 명령어
)
````

````python
clean_text_train2
````

````python
grid.fit(clean_text_train2, y_train)  # y_train으로 정답도 알려주기
````

````python
# 평가지표(정확도)
grid.best_score_
````

````python
# grid에 들어있는 속성값(경우의 수)
grid.best_params_
````

````python
# best model 옵션 가져오기
best_model = grid.best_estimator_
best_model
````

````python
# cross_val_score(lr_clf, trans_X_train, y_train, cv = 5).mean() 와 비교했을 때 하이퍼파라미터 모델이 미세하게 올랐음
best_model.score(clean_text_test2, y_test)
````

#### 학습결과 시각화
- 모델링 분석기법 중 가중치 분석
  - 어떠한 token(단어)이 혐오표현/악플/clean을 결정 짓는데 있어서, 영향을 많이 주었는가?

````python
# Pipeline에서 튜플(리스트 안에 소괄호) 형식으로 들어와있는 걸 확인할 수 있음
my_bm.steps
````

````python
# 인덱싱으로 확인 [(unmsile_cv , 객체) , (unsmile_logi, 객체)]에서 CountVectorizer 확인하고 싶다면?
# 각각의 독립적인 객체에 접근
cv = my_bm.steps[0][1]
# [(unmsile_cv , 객체) , (unsmile_logi, 객체)]에서 unsmile_logi 확인하고 싶다면?
logi = my_bm.steps[1][1]
````

````python
cv
````

````python
logi
````

````python
# 단어 사전 추출 - 우리 단어를 토큰화한 단어 사전(오른쪽 숫자들은 열번호)
voca = cv.vocabulary_
voca
````

````python
# 가중치 추출 - 어떠한 단어들이 악플에 영향을 많이 주었는지 확인하기
weight = logi.coef_[0]   # 혐오표현에 클래스에 대한 가중치 값들 - 다른 걸 확인하고 싶으면 인덱스 안에 번호 바꿔주기
weight
````

##### 그래프로 그려보기

````python
# 데이터프레임 만들기
unsmile_df = pd.DataFrame([voca.keys(), voca.values()]).T   # 단어사전에 들어있던 key값과 value값으로 DF 생성
unsmile_df.columns = ["단어", "인덱스"]   # 컬럼명 지어주기
````

````python
unsmile_df
# => 단어: 각각의 키값(단어사전) , 인덱스: 열번호 의미함
````

````python
# 인덱스 기준으로 정렬시키기
unsmile_df.sort_values(by = "인덱스")   
# => 가중치값을 맵핑시킬 수 있음
````

````python
unsmile_df.sort_values(by = "인덱스", inplace = True)   # 실제 인덱스 번호가 중구난방이기 때문에, 오름차순 정렬
````

````python
unsmile_df
````

````python
# 진짜 인덱스 번호를 왼쪽의 가짜 인덱스 자리에 배치(업데이트시키겠다)
unsmile_df.set_index("인덱스", inplace = True)
````

````python
unsmile_df
````

````python
# 각각의 단어의 가중치값들 넣어주기
unsmile_df["가중치"] = weight   # 정렬된 상태에서 가중치 컬럼 추가
unsmile_df
````

````python
# 가중치 기준으로 오름차순으로 정렬 , ascending = False : 내림차순으로 정렬
unsmile_df.sort_values(by ="가중치", ascending = False, inplace = True)   # 차트에서 정렬 안됨 -> 가중치별로 오름차순 정렬
unsmile_df.head(20)
````

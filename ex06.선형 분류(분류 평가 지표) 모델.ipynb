### 이상치 탐지 기법
- IQR 기법 : IQR(Q3-Q1)을 활용해서 이상치를 판단하는 기법
- 이상치 시준으로 만들어서 확인하는 것 -> 이상치인지 아닌지는 개발자의 몫!

```python
Q1 = wine['residual sugar'].quantile(0.25) # descibe() 에서의 25%
Q3 = wine['residual sugar'].quantile(0.75) # descibe() 에서의 75%

IQR = Q3 - Q1  #Inter Quartile Range, IQR (사분위수 범위)
```

```python
under = Q1 - (IQR * 1.5) #하한선을 찾은 거? 이 값을 넘어가는 수치는 다 이상치라고 판단하는 기법
over = Q3 + (IQR * 1.5) # 상한선
```

```python
# and, or : 논리연산자, 단일 값에 대해서 연산
# &, | : 비트 연산자, 다중 값에 대해서 연산
```

```python
wine[wine["residual sugar"] <= under]
# 하한선 밑으로는 없다.. 이상치로 판단할 데이터가 없다.

# 다중 불리언 인덱싱
wine[(wine["residual sugar"] <= under) | (wine['residual sugar'] >= over)]
# 둘 중 하나라도 틀리면 이상치
# 근데 무조건 이상치야! 라고 하는 건 아니고 그 컬럼들을 살펴봐야함. 
# redidual sugar가 거의 높은데 엥간 0 (레드 와인) 이니까 이상치가 아닐 수도 있다를 인지해야함. 
# 설탕이 높을 수록 0일 확률이 높겠구나! 느낌
```

```python
# 학습 준비
X = wine.iloc[:, :-1] # 문제
y = wine.iloc[:, -1]  # 정답
```

```python
# train, test 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size = 0.3,
                                                   random_state = 2024,
                                                   stratify = y)
# straify = y : y의 클래스 비율에 따라 데이터 분할, 분류 모델링시에만 수행
# 클래스 불균형 방지!
```

### Logistic Regression
- 분류만 가능!
- sigmoid(Logistic) 함수를 활용하여 0~1 사이의 확률값으로 예측하는 알고리즘
- 오차 계산할 때 사용되는 손실 함수는 Cross Entropy

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_clf = LogisticRegression(
    max_iter = 1000,
    C = 0.1, # 규제 강도, 숫자가 커질 수록 규제가 약하게 적용
    penalty = 'l2' # 규제 방법, Lasso
)
```

```python
# 학습
lr_clf.fit(X_train, y_train)
```

```python
# 예측
y_pred = lr_clf.predict(X_test)
y_pred
```

```python
# 평가
lr_clf.score(X_test, y_test)
```

#### 분류 평가 지표
| 혼돈 행렬 | 예측 | 실제 |
|-----------|------|------|
| TP(True Positive) | Positive | Positive |
| FP(False Positive) | Positive | Negative |
| TN(True Nagative) | Negative | Nagative |
| FN(False Nagative) | Negative | Positive |
---
- accuracy (정확도) : (TP + TN) / (TP+FP+TN+FN)
  - 전체 중에 맞춘(예측을 성공한) 비율(정확도)
  - 일반적으로 높은 정확도를 가진 모델을 선호
---
- racall(재현율) : TP / (TP + FN)
  - 실제 Positive인 것들 중에서 Positive로 올바르게 예측한 샘플의 비율
  - 긍정인 것들을 얼마나 잘 걸러내는지를 나타냄
  - 재현율이 높다 => 실제 Positive를 놓치지 않고 잘 포착한다는 의미를 가짐 (긍정인 것들을 잘 걸러낸다)
--- 
- precision(정밀도) : TP / (TP + FP)
  - Positive로 예측한 샘플들 중에서 실제 Positive인 샘플의 비율 (긍정이라고 예측한 것들 중에서 맞춘 것)
  - 정밀하게 잘 맞췄냐 => 우리 모델이 얼마나 적은 수의 FP(긍정으로 예측했는데 틀린 거)의 오류를 만들어냈는지를 의미
  - 정밀도가 높다 => 정밀하게 잘 맞춘다! (긍/부정을 잘 나눈다?)
---
- f1_score : 정밀도와 재현율의 조화 평균
  - 두 지표가 한 쪽에 치우치치 않고, Positive 케이스를 정확하고 효율적으로 식별하는 정도를 나타냄

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
```

```python
# 정확도 맞춘 개수, 일반적인 지표
accuracy_score(y_test, y_pred)
```

```python
# 재현율, 정밀도, f1_score 확인
recall_score(y_test, y_pred), precision_score(y_test, y_pred), f1_score(y_test, y_pred)
```
